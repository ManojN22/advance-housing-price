```python
def missing(feature,xtrain):
    xtrain[feature]=tat[feature]
   
    atain=pd.notnull(xtrain[feature])
    
    ates=pd.isnull(xtrain[feature])
    
    atrain=xtrain[atain]
    
    atest=xtrain[ates]
    
    trainid=atrain['index']
    
    testid=atest['index']
    
    btrain=atrain[feature]
    
    atrain=atrain.drop([feature,'index'],axis=1)
    
    atest=atest.drop([feature,'index'],axis=1)
    
    cl.fit(atrain,btrain)
    
    atest[feature]=cl.predict(atest)
    
    atest['index']=testid

    atrain['index']=trainid
    atrain[feature]=btrain
    
    xtrain=pd.concat([atrain,atest],axis=0)
    
    xtrain=xtrain.sort_values('index')
    xtrain=xtrain.drop('index',axis=1)
    xtrain=xtrain.reset_index()
    return xtrain
    
```

```python
def catonehot(fields,full):
    df_final=full
    
    df1=pd.get_dummies(full[fields],drop_first=True)  
    full.drop([fields],axis=1,inplace=True)
    
    for i in df1.columns:
        full[fields+i]=df1[i]
   
        
    return df_final
```
